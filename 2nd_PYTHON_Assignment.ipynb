```python
# Q30. Display students scoring more than 80

students = {
    "Satya": 85,
    "Rajat": 92,
    "Aditya": 70,
    "Kutun": 88,
    "Deepak": 75
}

print("Students scoring more than 80:")

for name, marks in students.items():
    if marks > 80:
        print(name, "-", marks)
```
